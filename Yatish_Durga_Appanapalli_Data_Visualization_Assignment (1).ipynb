### Task 0
Import numpy, pandas, matplotlib, seanborn, plotly, cufflinks

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected = True)
import plotly.express as px
import cufflinks as cf
cf.go_offline()
%matplotlib inline
```

### Task 1
Generate a one dimensional array x with integers from 0 to 99. Generate arrays for `y = 2x` and $z = x^{2}$

```python
array_x = np.arange(0, 100)
array_y = 2*np.array(array_x)
array_z = np.power(np.array(array_x), 2)
print("Array X: \n", array_x, "\n\nArray Y: \n", array_y, "\n\nArray Z: \n", array_z)
```

### Task 2
Create a figure object called fig using `plt.figure()`
Use `add_axes` to add an axis to the figure canvas at `[0,0,1,1]`. Call this new axis ax. Plot `(x,y)` on that axes and set the labels and titles to match the plot below:

```python
task2_fig = plt.figure()
t2_ax = task2_fig.add_axes([0,0,1,1])
t2_ax.grid(False)
t2_ax.set_xlabel('x')
t2_ax.set_ylabel('y')
t2_ax.plot(array_x,array_y)
t2_ax.set_title('My First Figure')
```

### Task 3
Create a figure object and put two axes on it, ax1 and ax2. Located at `[0,0,1,1]` and `[0.2,0.5,0.2,0.2]` respectively. Plot `(x,y)` on both axes.

```python
task3_fig = plt.figure()
t3_ax1 = task3_fig.add_axes([0,0,1,1])
t3_ax2 = task3_fig.add_axes([0.2,0.5,0.2,0.2])
t3_ax1.grid(False)
t3_ax2.grid(False)
t3_ax1.plot(array_x,array_y)
t3_ax2.plot(array_x,array_y)
t3_ax1.set_xlabel('x')
t3_ax1.set_ylabel('y')
t3_ax2.set_xlabel('x')
t3_ax2.set_ylabel('y')
```

### Task 4
Create the plot by adding two axes to a figure object at `[0,0,1,1]` and `[0.2,0.5,0.4,0.4]`. use x,y, and z arrays to recreate the plot below. Notice the x limits and y limits on the inner plot

```python
task4_fig = plt.figure()

t4_ax1 = task4_fig.add_axes([0,0,1,1])
t4_ax2 = task4_fig.add_axes([0.2,0.5,0.4,0.4])

t4_ax1.grid(False)
t4_ax2.grid(False)

t4_ax1.plot(array_x,array_z)
t4_ax1.set_xlabel('X')
t4_ax1.set_ylabel('Z')

t4_ax2.plot(array_x,array_y)
t4_ax2.set_title('Zoom In')
t4_ax2.set_xlabel('X')
t4_ax2.set_ylabel('Y')
t4_ax2.set_xlim([20, 25])
t4_ax2.set_ylim([30, 60])
```

### Task 5
Use `plt.subplots(nrows=1, ncols=2)` to create two subplots. Plot `(x,y)` and `(x,z)` on the axes. Play around with the linewidth and style

```python
t5fig, t5ax = plt.subplots(nrows=1, ncols=2)
t5ax[0].grid(False)
t5ax[1].grid(False)
t5ax[0].plot(array_x, array_y, 'b',linestyle = 'dashed',linewidth=3)
t5ax[1].plot(array_x, array_z, 'r', linewidth=3)
```

### Task 11
Create a Choropleth Plot of the Power Consumption for Countries using the data and layout dictionary.

```python
fig = px.choropleth(df, locations = 'Country', color = 'Power Consumption KWH', 
                    locationmode = 'country names',                 
                    color_continuous_scale = "ylorrd",
                    scope = "world", labels = {'Power Consumption KWH':'Power Consumption KWH'}, 
                    hover_name = 'Country',
                    title = 'Power Consumption in KWH')
fig.update_layout(margin = {"r":0,"t":0,"l":0,"b":0})
fig.update_geos(projection_type="natural earth")
fig.show()
```
